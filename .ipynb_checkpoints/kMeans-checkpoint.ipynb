```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from random import randint

x = []
y = []
for i in range(10):
    x.append(i + 1)
    y.append((x[i] + randint(-2,2)) * randint(1,2))

x = np.array(x)
y = np.array(y)

a, b, r, p, e = stats.linregress(x, y)
    
plt.plot(x, y, 'o')
plt.plot(x, a * x + b, 'r')
for i in range(10):
    plt.plot([x[i], x[i]], [y[i], a * x[i] + b], '-k')
plt.savefig('linreg.eps', format='eps')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

n_samples = 1500
random_state = 7
X, y = make_blobs(n_samples=n_samples, random_state=random_state)
colors = ['r', 'g', 'b']

plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], marker='.', c='k')
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.savefig('frame-0.eps', format='eps')
plt.show()
```

```python
from random import randint

C = []

for i in range(3):
    C.append([randint(-12, 12), randint(-3, 8)])

C = np.array(C)

plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], marker='.', c='k')
plt.scatter(C[:, 0], C[:, 1], 300, marker='D', c=['r', 'g', 'b'], edgecolor='k', linewidth=2)
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.savefig('frame-1.eps', format='eps')
plt.show()
```

```python
import math

y = y.astype('str')

for i in range(len(X)):
    d = math.inf
    for j in range(3):
        t = (X[i, 0] - C[j, 0]) ** 2 + (X[i, 1] - C[j, 1]) ** 2
        t = t ** 0.5
        if  t < d:
            y[i] = colors[j]
            d = t
            
plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], marker='.', c=y)
plt.scatter(C[:, 0], C[:, 1], 300, marker='D', c=['r', 'g', 'b'], edgecolor='k', linewidth=2)
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.savefig('frame-2.eps', format='eps')
plt.show()
```

```python
C[0] = [
    X[np.where(y == 'r'), 0].mean(), 
    X[np.where(y == 'r'), 1].mean()
]

C[1] = [
    X[np.where(y == 'g'), 0].mean(), 
    X[np.where(y == 'g'), 1].mean()
]

C[2] = [
    X[np.where(y == 'b'), 0].mean(), 
    X[np.where(y == 'b'), 1].mean()
]

plt.figure(figsize=(10, 8))
plt.scatter(X[:, 0], X[:, 1], marker='.', c=y)
plt.scatter(C[:, 0], C[:, 1], 300, marker='D', c=['r', 'g', 'b'], edgecolor='k', linewidth=2)
plt.xlabel("$X$")
plt.ylabel("$Y$")
#plt.savefig('frame-3.eps', format='eps')
plt.show()
```
